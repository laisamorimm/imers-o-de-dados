```python
# Creat one column

dados_resultados['n_moa'] = (dados_resultados.drop('id', axis = 1).sum(axis = 1))
```

```python
dados_resultados
```

```python
# Challenge: find the top 10 MOA (ihibitor,agonist)

dados_resultados.drop(['id','n_moa','ativo_moa'], axis = 1).sum()
```

```python
split_moa = np.unique([col.split('_')[-1] for col in dados_resultados.drop(['id','n_moa','ativo_moa'],axis = 1).columns])

frequencia = dados_resultados.drop(['id','n_moa','ativo_moa'], axis = 1).sum()

contadores = dict.fromkeys(split_moa,[0])
for name in frequencia.index:
    contadores[name.split('_')[-1]] += frequencia[name]
    
contagem_final = pd.DataFrame.from_dict(contadores).sum().sort_values(ascending = False).reset_index()
contagem_final.columns = ['ação', 'quantidade']
contagem_10 = contagem_final.head(10)
contagem_10
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x ,y ,test_size = 0.2, stratify = y, random_state = 376)

teste = []
treino = []

for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino,y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino,y_treino))
    
```
